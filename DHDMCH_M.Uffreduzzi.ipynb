## 3. Processamento dei dati

Procediamo allora con la pulizia dei dati.

Ma prima assicuriamoci anche che non ci siano duplicati.

```python
df.duplicated()
print(df.duplicated().value_counts())
```

Ok. Nessun duplicato.

Comincio sostituendo le lettere maiuscole presenti tra i valori di tipo stringa, così da uniformare i miei dati.

Verifico poi con i primi cinque valori.


```python
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].astype(str).str.lower()
```

```python
df.head()
```

Convertiamo ora i valori della colonna `data_creazione` in valori numerici e inseriamoli in una nuova colonna `'years'`.

Procediamo poi eliminando la colonna `data_creazione`.

```python
df['years'] = pd.to_numeric(df['data_creazione'], errors='coerce')
```

```python
df = df.drop(columns=['data_creazione'])
```

## 4. Analisi

Per iniziare la nostra analisi dobbiamo anzitutto farci un'idea della distribuzione delle opere nell'intervallo di tempo preso in considerazione nel nostro dataset.

Dunque isoliamo gli estremi e visualizziamo tramite grafico a linee.

```python
min_year = df['years'].min()
max_year = df['years'].max()

print(f"L'anno minimo presente nel DataFrame è: {min_year}")
print(f"L'anno massimo presente nel DataFrame è: {max_year}")
```

```python
opere_per_anno = df.groupby('years').size()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 8))
sns.lineplot(x=opere_per_anno.index, y=opere_per_anno.values, marker='o', linestyle='-')
plt.title('Numero di opere prodotte per anno')
plt.xlabel('Anno')
plt.ylabel('Numero di opere')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Un grafico di questo tipo ci permette di visualizzare immediatamente i dati sui quali stiamo lavorando. Riusciamo infatti ad individuare l'intero arco temporale con i relativi picchi al suo interno.

Il 1600 risulta essere senza dubbio l'anno più prolifico, seguito dal 1700 e dal 1500.

Molto probabilmente questo dato è dovuto agli arrotondamenti dell'archivistica, che spesso assegna cifra tonda in caso di datazione non certa.



```python
anni_piu_prolifici = opere_per_anno.sort_values(ascending=False)
display(anni_piu_prolifici.head(10))
```

Concentriamo ora la nostra ricerca restringendo il campo al genere del ritratto.

```python
df_ritratto = df[df['generi'].str.contains('ritratto', na=False)]
opere_ritratto_per_anno = df_ritratto.groupby('years').size()

plt.figure(figsize=(15, 7))
plt.scatter(x=opere_ritratto_per_anno.index, y=opere_ritratto_per_anno.values, color='red', alpha=0.7)
plt.title("Produzione di opere di genere 'ritratto' per anno (in scatter_plot)")
plt.xlabel('Anno')
plt.ylabel('Numero di opere di genere ritratto')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
anni_piu_prolifici_ritratto = opere_ritratto_per_anno.sort_values(ascending=False)
display(anni_piu_prolifici_ritratto.head(10))
```

In seguito alla manipolazione e all'analisi di questi dati abbiamo potuto constatare anzitutto i possibili *bias* presenti in un dataset. Le problematiche possono essere preesistenti e dovute alla catalogazione, come ad esempio la datazione approssimativa a inizio secolo, o l'incompletezza dei dati, oppure dovute all'importazione stessa.

La ricerca ha evidenzato il 1600 come anno più prolifico per l'arte pittorica in generale; mentre il 1900 e il 1545 sono a parimerito per la produzione di ritratti.

Va però ricordato che è emerso un valore nan non indifferente. Inoltre, confrontando i risultati emersi con le conoscenze storico-artistiche pregresse, sono state individuate importanti lacune nel dataframe, che avrebbero sicuramente influito sui risultati della ricerca.
